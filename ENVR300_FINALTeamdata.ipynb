```r
#load packages
library(tidyverse)
library(here)
library(tidymodels)
library(gtools)
```

```r
#the function to use for all gcms
#add col for gcm name and remove NA values
my_fun <- function(x) {
    x%>%
        select("NaN.","X1","X2","X6","X7","X8","X12", "gcm") %>%   #get summer and winter months only
        na.omit()                                                  #get rid of NA values
                }
```

# GCM: CanESM2 #

```r
#sort files for a gcm
canesm_files <- list.files(pattern="T_CanESM2")
canesm_files <- mixedsort(canesm_files) #make sure files r listed in order

##Create character vector of df names without the ".csv" part 
canesm_names <-substr(canesm_files,3,8)

#rename files neatly <gsm><#>
for (i in 1:10) {
    canesm_names[i] <- paste("canesm",i,sep="")}

###Load all files as df
for(i in 1:10){
    filepath <- file.path(canesm_files[i],paste(i,sep=""))
    assign(canesm_names[i], read.csv(toString(canesm_files[i]), 
        ))
}

#make list of dfs
canesm.list <- list(canesm1, canesm2, canesm3, canesm4, canesm5, canesm6, canesm7, canesm8, canesm9, canesm10)

#apply function
canesm.list2 <- lapply(canesm.list, function(x) {mutate(x, gcm = 'canesm')}) #add col for gcm name
canesm.list3 <- lapply(canesm.list2, my_fun)
```

# Data for Each Station #

```r
#make list with 10 empty dfs for each station's data
station <- replicate(10, data.frame())
print(station[[1]])
```

```r
#create a dataframe for each station, merging gcms
for (i in 1:10) {
    station[[i]] <- rbind(                     #merge station i from each gcm into a dataframe
        as.data.frame(canesm.list3[i]), 
        as.data.frame(csiro.list3[i]), 
        as.data.frame(gfdl.list3[i]), 
        as.data.frame(miroc.list3[i]), 
        as.data.frame(mpi.list3[i]))
     station[[i]] <- rename(station[[i]],       #rename cols
            `Year`=NaN.,
            `1`=X1,
            `2`=X2,
            `6`=X6,
            `7`=X7,
            `8`=X8,
            `12`=X12)
        station[[i]] <- station[[i]] %>%    #add new col for means for winter and summer
                    mutate(winter = rowMeans(station[[i]][,c("1", "2", "12")]),
                          summer = rowMeans(station[[i]][,c("6","7","8")]))    
    }

station[[10]]
```

# Make 'em plots - station 10 winter #

```r
plot1 <- station[[10]] %>% 
    ggplot() + 
    aes(x = Year, y = winter, col = gcm) +
    geom_point(size=0.5) +
    geom_smooth(method = "lm", se=FALSE) +
    labs(title="Temperature vs Year for Station 10 in Winter",
         x="Year",
         y="Temperature (oC)") +
    scale_color_discrete(name="Global Climate Model",
                         labels=c("CanESM2","CSIRO-Mk3.6.0", "GFDL-CM3","MIROC-ESM","MPI-ESM-LR"))+
    theme(text = element_text(size=12))

plot1
```

```r
plot2 <- station[[10]] %>% 
    ggplot() + 
    aes(x = Year, y = summer, col = gcm) +
    geom_point(size=0.5) +
    geom_smooth(method = "lm", se=FALSE) +
    labs(title="Temperature vs Year for Station 10 in Summer",
         x="Year",
         y="Temperature (oC)") +
    scale_color_discrete(name="Global Climate Model",
                         labels=c("CanESM2","CSIRO-Mk3.6.0", "GFDL-CM3","MIROC-ESM","MPI-ESM-LR"))+
    theme(text = element_text(size=12))

plot2
```

# Get linear regression data #

```r
#name of cols
gcmnames <- c("Station", "CanESM2","CSIRO-Mk3.6.0", "GFDL-CM3","MIROC-ESM","MPI-ESM-LR")

#empty df to store slope values WINTER
finalvals_w <- data.frame(matrix(nrow=10, ncol=6)) #empty df with 10rows and 6 cols
colnames(finalvals_w) = gcmnames #rename cols
finalvals_w$Station <- 1:10 #insert numbers for Station col

#empty df to store slope values SUMMER

finalvals_s <- data.frame(matrix(nrow=10, ncol=6)) #empty df with 10rows and 6 cols
colnames(finalvals_s) = gcmnames #rename cols
finalvals_s$Station <- 1:10 #insert numbers for Station col

finalvals_s
```

```r
#loop thru stations and gcms collecting slope values and adding to finalvals_w df

gcmnames <- c("canesm","csiro", "gfdl","miroc","mpi")

for (z in 1:10) {      #loop thru stations
    for (y in 1:5) {   #loop thru gcms
        tempdata = subset(station[[z]], gcm == gcmnames[y])   #subset only 1 gcm, 1 station
        slope = coef(lm(winter~Year, data=tempdata))[2]       #get slope value
        finalvals_w[z,(y+1)] <- slope                         #add slope value to finalvals_w df
        finalvals_w$Difference[z] <- max(finalvals_w[z,2:6]) - min(finalvals_w[z,2:6])
        }
    }

finalvals_w
```

```r
#loop thru stations and gcms collecting slope values and adding to finalvals_w df

for (z in 1:10) {      #loop thru stations
    for (y in 1:5) {   #loop thru gcms
        tempdata = subset(station[[z]], gcm == gcmnames[y])   #subset only 1 gcm, 1 station
        slope = coef(lm(summer~Year, data=tempdata))[2]       #get slope value
        finalvals_s[z,(y+1)] <- slope                         #add slope value to finalvals_w df
        
        finalvals_s$Difference[z] <- max(finalvals_s[z,2:6]) - min(finalvals_s[z,2:6])   #calculate difference
        }
    }

finalvals_s
```

```r
#reorder dfs to rank
ranking_w <- finalvals_w[order(-finalvals_w$"Difference"),]
ranking_w

ranking_w <- finalvals_w[order(-finalvals_w$"Difference"),]
ranking_w
```
